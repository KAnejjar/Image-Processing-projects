### 1.Getting familiar with image manipulation

```python
#importing the libs 
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

##### (a)	Read an image, convert it to grayscale if it isnâ€™t already, and display the converted image

```python
img = cv2.imread('images/lion.jpg')
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
plt.title('Gray Lion')
plt.show()
```

#### (b)	Calculate the size (total number of pixels) of the image

```python
height = img.shape[0]
width = img.shape[1]
total_pixels = width*height
print("Width => {}\nHeight => {}\nTotal number of pixels => {} pixels".format(width,height,total_pixels))
```

#### (c)	Calculate the maximum pixel value 

```python
#max for R channel,for G channel and for B channel
#the result is the max value of them all
max_channels = np.amax([np.amax(img[:,:,0]), np.amax(img[:,:,1]), np.amax(img[:,:,2])])
print("Maximum pixel value => ",max_channels)
#white pixel
```

#### (d)	Calculate the mean pixel value

```python
image = img[:,:,:]
mean = image.mean()
print("mean pixel value ==>", mean)
```

#### (e)	Change the pixel values of the image in the following way: all pixelsâ€™ values less than the average calculated at (d) will be equal to 0 and all the others will be equal to 1.   Make sure your displayed image is black and white, not black and near black.

```python
new_image = cv2.imread('images/lion.jpg')
print("Loading..")
for i in range(width):
    for j in range(height):
        mean_pixel = img[i][j].mean()
        if mean_pixel<mean:
            new_image[i][j]=[0,0,0]
        else:
            new_image[i][j]=[255,255,255]

plt.imshow(new_image,cmap="gray")
plt.title('Black and White Lion')
plt.show()      
```

### 2.	Image Interpolation 

##### Write a computer program that will, given an input image, reduce its spatial resolution,and then return it to its original resolution.  Use all of nearest neighbor, bilinear and bicubic interpolation to do this. Your display should include at least 7 images: the original, three reduced images and three restored images.  Comment on the differences you see among the three restored images.  

```python
plt.imshow(rgb_img)
plt.title('Original Image')
plt.show()

# Downscaling the image
new_height = int(height/10)
new_width = int(width/10)
 
# Performing Nearest neighbor interpolation
nni_image = cv2.resize(rgb_img, (new_width, new_height), interpolation = cv2.INTER_NEAREST)
restored_nni_image = cv2.resize(nni_image, (new_height*10, new_width*10), interpolation = cv2.INTER_NEAREST)
#restored_nni_image = cv2.GaussianBlur(restored_nni_image,(7,7),0)

# Performing bilinear interpolation
bii_image = cv2.resize(rgb_img, (new_width, new_height), interpolation = cv2.INTER_LINEAR)
restored_bii_image = cv2.resize(bii_image, (new_height*10, new_width*10), interpolation = cv2.INTER_LINEAR)
#restored_bii_image = cv2.GaussianBlur(restored_bii_image,(7,7),0)

# Performing bicubic interpolation
bci_image = cv2.resize(rgb_img, (new_width, new_height), interpolation = cv2.INTER_CUBIC)
restored_bci_image = cv2.resize(bci_image, (new_height*10, new_width*10), interpolation = cv2.INTER_CUBIC)
#restored_bci_image = cv2.GaussianBlur(restored_bci_image,(7,7),0)

# Displaying the results
fig = plt.figure(figsize=(20, 14))
rows = 3
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(nni_image)
plt.title("Nearest neighbor interpolation")

fig.add_subplot(rows, columns, 2)
plt.imshow(restored_nni_image)
plt.title("Restored Nearest neighbor interpolation")

fig.add_subplot(rows, columns, 3)
plt.imshow(nni_image)
plt.title("bilinear interpolation")

fig.add_subplot(rows, columns, 4)
plt.imshow(nni_image)
plt.title("Restored bilinear interpolation")

fig.add_subplot(rows, columns, 5)
plt.imshow(nni_image)
plt.title("bicubic interpolation")

fig.add_subplot(rows, columns, 6)
plt.imshow(nni_image)
plt.title("Restored bicubic interpolation")
```
