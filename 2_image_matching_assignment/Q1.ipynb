```python
from PIL import Image
```

```python
def split_image(src_img,identifier,rows=3,cols=3):
    im = Image.open(src_img)
    im_w, im_h = im.size
    w = int(im_w/cols)
    h = int(im_h/rows)
    w_num, h_num = int(im_w/w), int(im_h/h)

    for wi in range(0, w_num):
        for hi in range(0, h_num):
            box = (wi*w, hi*h, (wi+1)*w, (hi+1)*h)
            piece = im.crop(box)
            tmp_img = Image.new('RGB', (w, h), 255)
            tmp_img.paste(piece)
         
            img_path = "output_images\\{}_{}{}.jpg".format(identifier,wi,hi)
            print(img_path)
            tmp_img.save(img_path)
```
